```python
name = input("Enter your name: ")
Sub1 = int(input("Enter maths marks: "))
Sub2 = int(input("Enter english marks: "))
Sub3 = int(input("Enter science marks: "))
Sub4 = int(input("Enter chemistry marks: "))
Sub5 = int(input("Enter biology marks: "))
total_marks = Sub1 + Sub2 + Sub3 + Sub4 + Sub5
max_marks = 500

percentage = (total_marks / max_marks) * 100

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
else:
    grade = "F"

print("Name:", name)
print("Math:", Sub1)
print("English:", Sub2)
print("Science:", Sub3)
print("Chemistry:", Sub4)
print("Biology:", Sub5)
print("Total:", total_marks, "/", max_marks)
print("Percentage:", round(percentage, 1), "%")
print("Grade:", grade)
```
